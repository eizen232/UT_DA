```python
# Define bins as 0-25, 25-50, 50-75, 75-100 (note that 25 will be in 1st bucket, 50 will be in 2nd bucket (i.e. they won't be present in both buckets))

bins = [0, 25, 50, 75, 100]

# name the three groups

group_names = ['young', 'middle', 'old', 'older']

# use pd.cut to create the bins and the new column

credit_idrmv['age_group'] = pd.cut(credit_idrmv['AGE'], bins=bins, labels = group_names)

```

```python
credit_idrmv.head()
```

```python
# Define bins as 0-200k, 200k-400k, 400k-600k, 600k-800k, 800k-1M (note that 25 will be in 1st bucket, 50 will be in 2nd bucket (i.e. they won't be present in both buckets))

bins = [0, 200000, 400000, 600000, 800000, 1000000]

# name the three groups

group_names = ['0-200k', '200k-400k', '400k-600k', '600k-800k', '800k-1M']

# use pd.cut to create the bins and the new column

credit_idrmv['loanbal_grp'] = pd.cut(credit_idrmv['LIMIT_BAL'], bins=bins, labels = group_names)
```

```python
# Create a function to define those who are young (less than 25)
def male_female_child(borrower):
    age, sex = borrower
    
    if age < 25:
        return 3
    else:
        return sex
    
credit_idrmv_test['person'] = credit_idrmv[['AGE', 'SEX']].apply(male_female_child, axis=1) 
```
